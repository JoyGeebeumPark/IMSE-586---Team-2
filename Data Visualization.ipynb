# EXPLORATORY DATA ANALYSIS

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
insurance = pd.read_csv('./downloads/expenses.csv')
insurance.head()
```

```python
bar = insurance["smoker"].value_counts()
bar
```

```python
plt.figure(figsize=(10,20))

plt.subplot(4,2,1)
sns.countplot("smoker", data=insurance)
plt.xlabel("Smokers", fontsize =  12)
plt.ylabel("Value counts of Smokers", fontsize =  12)

plt.subplot(4,2,2)
sns.countplot("region", data=insurance)
plt.xlabel("Region", fontsize =  12)
plt.ylabel("Value counts for Region", fontsize =  12)

plt.subplot(4,2,3)
plt.hist(insurance["age"], edgecolor="k")
plt.xlabel("Age", fontsize =  12)

plt.subplot(4,2,4)
sns.countplot("children", data=insurance)
plt.xlabel("Children", fontsize =  12)
plt.ylabel("Value counts for Children", fontsize =  12)


plt.subplot(4,2,5)
sns.countplot("sex", data=insurance)
plt.xlabel("Sex", fontsize =  12)
plt.ylabel("Value counts for Sex", fontsize =  12)

plt.subplot(4,2,6)
plt.hist(insurance["bmi"], edgecolor="k")
plt.xlabel("Bmi", fontsize =  12)

plt.subplot(4,2,7)
plt.hist(insurance["charges"], edgecolor="k")
plt.xlabel("Charges", fontsize =  12)

plt.show()
```

```python
plt.figure(figsize=(20,10))

plt.subplot(2,2,1)
sns.regplot(x="age", y="charges", data=insurance, marker="o")
plt.xlabel("Age", fontsize =  16)
plt.ylabel("Charges", fontsize = 16)
plt.title("Scatter plot of Age vs Charges", fontsize = 12)

plt.subplot(2,2,2)
sns.regplot(x="bmi", y="charges", data=insurance, marker="o")
plt.xlabel("Bmi", fontsize =  16)
plt.ylabel("Charges", fontsize = 16)
plt.title("Scatter plot of Bmi vs Charges", fontsize = 12)

plt.show()
```

```python
plt.figure(figsize=(20,20))

plt.subplot(4,2,1)
sns.swarmplot(x="sex", y="charges", data=insurance, alpha=0.7)
sns.boxplot(x="sex", y="charges", data=insurance, color="white", fliersize=0)
plt.xlabel("Sex", fontsize =  16)
plt.ylabel("Charges", fontsize = 16)
plt.title("Swarm plot of Sex", fontsize = 12)

plt.subplot(4,2,2)
sns.swarmplot(x="smoker", y="charges", data=insurance, alpha=0.7)
sns.boxplot(x="smoker", y="charges", data=insurance, color="white", fliersize=0)
plt.xlabel("Smoker", fontsize =  16)
plt.ylabel("Charges", fontsize = 16)
plt.title("Swarm plot of Smokers")

plt.subplot(4,2,3)
sns.swarmplot(x="children", y="charges", data=insurance, alpha=0.7)
sns.boxplot(x="children", y="charges", data=insurance, color="white", fliersize=0)
plt.xlabel("Children", fontsize =  16)
plt.ylabel("Charges", fontsize = 16)
plt.title("Swarm plot for Children")

plt.subplot(4,2,4)
sns.swarmplot(x="region", y="charges", data=insurance, alpha=0.7)
sns.boxplot(x="region", y="charges", data=insurance, color="white", fliersize=0)
plt.xlabel("Region", fontsize =  16)
plt.ylabel("Charges", fontsize = 16)
plt.title("Swarm plot for Region")

plt.show()
```
